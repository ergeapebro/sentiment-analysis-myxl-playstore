```python
from google_play_scraper import reviews, Sort
import pandas as pd

app_id = "com.apps.MyXL"

# Batas jumlah review
max_reviews = 20000
count_per_request = 1000  # Ambil 100 review per request
all_reviews = []
next_token = None

while len(all_reviews) < max_reviews:
    count = min(count_per_request, max_reviews - len(all_reviews))
    
    result, next_token = reviews(
        app_id,
        lang="id",
        country="id",
        sort=Sort.MOST_RELEVANT, 
        count=count, 
        continuation_token=next_token
    )
    
    all_reviews.extend(result)
    
    if not next_token:
        break 

df = pd.DataFrame(all_reviews)

df.to_csv("myxl_reviews.csv", index=False, encoding="utf-8")

print(f"Scraping selesai! Total {len(df)} ulasan disimpan dalam 'myxl_reviews.csv'")
```
